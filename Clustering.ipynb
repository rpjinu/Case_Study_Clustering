
# A case study on Clustering


A supermarket called K-mart wants to analyse its customers. It wants to understand its customers, their spends, their marketing campaigns and make it easier for them to take business decisions and modify products and strategies accordingly.

This process is called customer personality analysis. When analysing customers, its also valuable to cluster the customers into meaningful clusters so that we can make our decisions accordingly and make customer cluster specific decisions. We want to be in a position where we can try different strategies for different customer clusters.
You are the official Data scientist of the K-mart and are given the task to do the customer personality analysis and also do the clustering of the customers.

# Clustering

A major step in working with the customers is segregating them in multiple clusters based on the data.
This seggregation is also called clustering. 

Clustering is a field of Machine learning which belongs to unsupervised learning. Unsupervised because we don't have labelled data, we are trying to do learning over unlabelled data hence no "supervision".

Clustering deals with putting the data points that are similar to each other  sinto meaningful clusters, in our case if we can segregate our customers properly in multiple clusters, we can apply our sales and marketing startegy accordingly.

There are many techniques on clustering. Most common ones being K- means clustering, DBSCAN and few more.
The simplest one being k-means.



## K-means

K-means work on a simple idea. Bring out the points that are closer to each other on the cartesian plane together and call it one cluster.

OK, dont worry Lets have a detailed look!

1. So we start by intialising k points and call them cluster centroids, randomly.
2,. Once that's done, for each point, we look at the closest mean and assign it that cluster. Basically for every point, we itemise a k (or a mean)
3. We then update the mean's coordinates, by taking the average of all the points that were assigned that cluster.
4. So now we have new cluster means, we repeat this process of assigning and updating cluster means and at the end we have our clusters.

Some points to notice.
> * The cluster mean updation stops the moment we stop seeing any updates to the means, i.e if means are not updating, that means we have found our clusters and we dont need to do k means now.
> * To find the distance between a point and its cluster mean, k-means uses cartesian distance between 2 points.


## Finding the K in K-Means

OK now that we have understood the k means clustering. There is still one question that lies ahead of us. How do we estimate the k in k means clustering. We use a very nice technique to estimate the k in k-means clusteing, its called Elbow.


### Elbow method


Elbow gives us a graphical way of finding the optimal 'K' in k means clustering. 
It calcluates a number called WCSS or within cluster sum of square.WCSS is basically computed by finding the distance of each point in given cluster from its centroid. The idea is that if we are making good clusters, we should be able to reduce this score i.e clsuters' points should be closer to the centroid. WCSS is also called distortion score.
Elbow method is a grahpical method, we compute the value of WCSS or distortion score and plot them against many K values. The value of k where the score stops decreasing significantly is called the elbow, because visually it does look like an elbow. 
The value of k where the elbow happens is the k we choose to use in the k-means clustering algorithm.

Don't worry, it will start to make more sense as we look at it visually.

### Silhouette Method

This is another method to estimate the k in k-means clustering.
It works on something called silhouette coefficient, which is a measure of similarity of a data point within the cluster in comparison to other clusters. 
The process is to calculate the silhouette coefficient for a range of k values and then plot them. 
$ S(i) = \frac{b(i)-a(i)}{max{\{a(i),b(i)\}}}  $

Where S(i) is the silhouette coefficient for a given data point i 
a(i) is average distance between ith data point and other points in the same cluster.
b(i) is the average distance between i and the other remaining clusters where i doensn't belong.

```python
from sklearn.metrics import silhouette_score
clusterRange = [2, 3, 4, 5, 6, 7, 8]
avgSilhouette = []
for numClusters in clusterRange:
 
 # initialise kmeans
 kmeans= KMeans(n_clusters=numClusters)
 kmeans.fit(newPCs)
 clusterLabels = kmeans.labels_
 avgSilhouette.append(silhouette_score(newPCs, clusterLabels))


 # silhouette score
plt.plot(clusterRange,avgSilhouette)
plt.xlabel('K') 
plt.ylabel('Silhouette score') 
plt.show()
```

As we can see the silhouette plot is peaking at k =4, but we want to further analyse the silhouette analysis, hence we will be doing a plot on silhouette analysis.

Silhouette analysis can be used to study the separation distance between the resulting clusters. The silhouette plot displays a measure of how close each point in one cluster is to points in the neighboring clusters and thus provides a way to assess parameters like number of clusters visually. This measure has a range of [-1, 1].

Silhouette coefficients (as these values are referred to as) near +1 indicate that the sample is far away from the neighboring clusters. A value of 0 indicates that the sample is on or very close to the decision boundary between two neighboring clusters and negative values indicate that those samples might have been assigned to the wrong cluster.


source: https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html

We also need to see how much business sense these clusters make. To understand the business meaning of these clusters.
We will be plotting the one of the most common variable that help us understand the customers effectively, the total expense they are doing at the store vs the unique customer id.

This helps us bucket our customers into proper clusters and make our marketing and campaigns strategy effectively. Lets plot our cluster with k=4 and also look at the how the clusters for the CombinedExpense column

```python
# consider the number of clusters
n_clusters = [2, 3, 4, 5, 6]

# consider an array of the data
X = newPCs

# for each value of K, plot the silhouette plot the clusters formed
for K in n_clusters:
    
    # create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    
    # set the figure size
    fig.set_size_inches(18, 7)

    # the 1st subplot is the silhouette plot
    # initialize the cluster with 'K' value and a random generator
    model = KMeans(n_clusters = K, random_state = 10)
    
    # fit and predict on the scaled data
    cluster_labels = model.fit_predict(X)

    # the 'silhouette_score()' gives the average value for all the samples
    silhouette_avg = silhouette_score(X, cluster_labels)
    
    # Compute the silhouette coefficient for each sample
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(K):
        
        # aggregate the silhouette scores for samples belonging to cluster i, and sort them
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
        
        # sort the silhouette coefficient
        ith_cluster_silhouette_values.sort()
        
        # calculate the size of the cluster
        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        # color each cluster 
        color = cm.nipy_spectral(float(i) / K)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # compute the new y_lower for next plot
        y_lower = y_upper + 10 

    # set the axes and plot label
    ax1.set_title("Silhouette Plot")
    ax1.set_xlabel("Silhouette coefficient")
    ax1.set_ylabel("Cluster label")

    # plot the vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    # clear the y-axis ticks
    ax1.set_yticks([])  
    
    # set the ticks for x-axis 
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8])

    

    # label the cluster centers

    newPCs['cluster']=cluster_labels
    
    

    fig.show()
    df['Cluster']=cluster_labels
    sns.scatterplot(x='ID',y='CombinedExpense', data=df, hue='Cluster')
    

    # add the axes and plot title
    ax2.set_title("Clusters")
    ax2.set_xlabel("Customer ID")
    ax2.set_ylabel("Combined Expense")
    
    # set the common title for subplots
    plt.suptitle(("Silhouette Analysis for K-Means Clustering with n_clusters = %d" % K), fontsize=14, 
                 fontweight='bold')

# display the plot
plt.show()
```

As we can see k=4, is making a lot of sense in the clusters again, also k=4 achieves maximum average silhouetter score.

# Hierarichal Clustering

This is another clustering technique, its a bit simpler than K means clustering. Initially every point in the dataset is considered a cluster in this techbnique.

The basic algorithm is straight forward.

1. Start by computing the proximity matrix
2. Initially, consider each data point a cluster.
3. Merge the 2 closest point into 1 l cluster and update the proximity matrix
4. Repeat this process until 1 cluster remains.


Now to measure this closeness or similarity between 2 clusters, we can use euclidean distanse as the matrix. 
Proximity matrix is nothing but a matrix that records this distance amongst the clusters and we update it post every iteration.
The type of clustering that we just understood is called Agglomerative clustering. Another type of hierarichal called Divisive Hierarchical clustering exists which is the opposite of the Agglomerative clustering. 

In this case study lets focus on the Agglomerative hierarichal clustering. 


#### Estimating the number of clusters in Agglomerative hierarichal clustering.
To estimate the number of clusters, in this case we use somethign called dendogram. 

```python
# import cluster hierarchy
import scipy.cluster.hierarchy as sch
#plot the dendogram
dendrogrm = sch.dendrogram(sch.linkage(newPCs, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Features')
plt.ylabel('Euclidean distance')
plt.show()
```

In dendogram, more the distance between vertical lines, more the distance between clusters. Dendogram represents the hierarichal clusteting process as well. If you look at the diagram bottom up, at the bottom you will see small clusters forming, i.e 1 point clusters, as you go up these box size increases forming even bigger clusters and biggest ones at the top.

To find out the optimal number of clusters, we draw a horinzontal line cutting the biggest vertical lines.
The number of vertical lines cut by our horizontal line is our number of clusters. 

In the case above, we could draw a line cutting either 2 blue vertical lines or a line cutting 1 blue and 2 red lines. Either choice giving us 2 or 3 clusters.

## DBSCAN or Density based clustering 

In K means and even Hierarhical clustering, there are 2 slight disadvantages:
1. You need to provide a k value
2. Each data point plays a role in forming the clusters. A slight change in data points might affect the clustering outcome

To overcome this, introducing Density based clustering or DBSCAN. 
DBSCAN doesn't need the number of clusters beforehand, and its also not extremely sensitive to the data points.




The DBSCAN  just needs two parameters:

minPts: The minimum number of points (a threshold) clustered together for a region to be considered dense.
eps (ε): A distance measure that will be used to locate the points in the neighborhood of any point.

Consider a set of points in some space to be clustered. Let ε be a parameter specifying the radius of a neighborhood with respect to some point. For the purpose of DBSCAN clustering, the points are classified as core points, directly-reachable points and outliers, as follows:

A point p is a **core point** if at least minPts points are within distance ε of it (including p).
A point q is **directly reachable** from p if point q is within distance ε from core point p. Points are only said to be directly reachable from core points.
A point q is reachable from p if there is a path p1, ..., pn with p1 = p and pn = q, where each pi+1 is directly reachable from pi. Note that this implies that the initial point and all points on the path must be core points, with the possible exception of q.
All points not reachable from any other point are **outliers** or noise points.


Now if p is a core point, then it forms a cluster together with all points (core or non-core) that are reachable from it. Each cluster contains at least one core point; non-core points can be part of a cluster, but they form its "edge", since they cannot be used to reach more point


_source: wikipedia_

Algorithmic steps for DBSCAN clustering


1. We start by arbitrarily picking up a point in the dataset, initially.
2. If we find there are at least 'minPoint' points within a radius of ‘ε’ (directly reachable points), then all such points are considered to be part of the same cluster.
3. This process is repeated recusively to create and expand clusters for every neighbouring point. 


#### Deriving parameters of DBSCAN:

1. minPts: Rule of thumb is to take this value greater than 2. Since minPts =1 wont make any sense , minPts=2 would give the same result as Hierarichal clustering. The general rule is minPts ≥ D + 1, where D is the dimesionality of the data.

2. ε can be estimated by the k-distance graph, plotting the distance to the k = minPts-1 nearest neighbor sorted in increasing order. We are looking for the ε that gives us the elbow. Too small ε would mean large part of data unclustered, too high value of ε would mean no or less number of clusters.
